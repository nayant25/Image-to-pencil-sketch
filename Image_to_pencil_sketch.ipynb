# Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

```python
from google.colab.patches import cv2_imshow
```

# Importing the image

```python
cat_image = cv2.imread("cat.jpg")
cv2_imshow(cat_image)
```

# Reading in RBG format 

```python
rbg = cv2.cvtColor(cat_image, cv2.COLOR_BGR2RGB)
cv2_imshow(rbg)
```

# Converting RBG format to grayscale

```python
gray_scale = cv2.cvtColor(rbg, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_scale)
```

# Inverted Gray scale

```python
inverted_gray_scale = 255 - gray_scale
cv2_imshow(inverted_gray_scale )
```

# Creating a blurred image

```python
blur = cv2.GaussianBlur(inverted_gray_scale, (31,31), 0)
cv2_imshow(blur)
```

# Inverted blur image

```python
inverted_blur = 255 - blur
cv2_imshow(inverted_blur)
```

# Creating a pencil sketch

```python
pencil_sketch = cv2.divide(gray_scale, inverted_blur, scale=255)
cv2_imshow(pencil_sketch)
```

# Original image and pencil sketch

```python
cv2_imshow(cat_image)
cv2_imshow(pencil_sketch)
```
